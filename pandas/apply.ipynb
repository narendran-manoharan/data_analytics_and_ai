## You are applying the function you created, to a particular column.

```python
def total (df):
    return df["x"] + df["y"] + df["z"]
df.loc [:, "total"] = df.apply(total, axis=1) #This is how you apply. You first create a new column. But updates recommend that you add a .loc and call out the entire row. Then you assign arguments - df.apply("apply the function you created", to the column by calling out axis 1).
df
```

```python
def product (df):
    return df["price"] * 4
df.loc[:, "product"] = df.apply(product, axis=1)
df
```

```python
df = sns.load_dataset("tips")
df
```

```python
def tips_proportion(df):
    return df["tip"] / df["total_bill"]
df.loc[:, "tips_proportion"] = df.apply(tips_proportion, axis=1)
df
```
