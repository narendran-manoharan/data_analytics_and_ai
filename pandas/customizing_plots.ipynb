```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("insurance.csv")
df= df.sample(n=5, random_state=9)
df
```

A bar plot showing the distribution of the number of children.
A pie plot showing the proportional count of different regions in the dataset.
A scatter plot describing the relationship between 'age' and 'charges'.

```python
fig, axes = plt.subplots(nrows= 3, ncols=3, figsize=(12, 8))
categorical_count= df.value_counts("children")
axes[0,1].bar(data= df, x=categorical_count.index, height=categorical_count)

categorical_count= df.value_counts("region", normalize=True)
axes[1,2].pie(x= categorical_count, labels= categorical_count.index)

axes[2, 2].scatter(x= df["age"], y=df["charges"], c=df["charges"])

plt.show()
```

### We will create a Figure with 3 Axes and draw a line plot in the first, and a histogram for y1 and y2 in the second and third Axes, respectively.

```python
x = np.linspace(3, 45, 40)
y1 = np.sin(x)
z= x * np.sin(x)

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12,8))
axes[0].plot(x, y1, label="sin(x)")
axes[0].plot(x, z, label="(x) times sin(x)")
axes[0].legend()

axes[1].hist(y1, label="sin(x)", bins=10)
axes[1].legend()

axes[2].hist(z, label= "(x) times sin(x)", bins= 10)
axes[2].legend()

plt.show()
```

```python
def Mulitpleplots():
    my_var = np.linspace(0, 555, 555)

    fig, axes= plt.subplots(nrows=3, ncols=3, figsize=(10,8))
    axes[0, 1].plot(my_var, np.sin(my_var))
    axes[0,1].set_xlabel("X-axis")
    axes[0,1].set_ylabel("Y-axis", rotation=None)
    axes[0, 1].set_title("Plot 1")
    axes[1, 2].plot(my_var, my_var*np.sin(my_var))
    axes[2, 0].plot(my_var, my_var*np.sin(my_var ** 2))
    axes[2, 2].plot(my_var, np.cos(2*my_var) * np.sin(my_var))

    return fig, axes

fig, axes = Mulitpleplots()
fig.suptitle("My First Trial", fontsize= 18, y=1)
plt.tight_layout()
plt.show()   
```

You can add and customise your legend

Consider 2 arrays plotted in the same Axes
The first is a mathematical function of sin(x), and the second is x * sin(x)
You will notice the argument label at plt.plot(), where you set how you want to call that particular plot.
You will notice the labels are not displaying.

```python
df = np.linspace(0, 10, 500)
x1= np.sin(df)
y1= x1 * np.sin(df)

plt.plot(df, x1, label="Axes 1")
plt.plot(df, y1, label="Axes 2")
plt.legend(loc="best", title = "Legend", frameon= False)


plt.suptitle("Extra")
plt.show()
```
