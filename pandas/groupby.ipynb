```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame({"Product": ["Bread", "Bread", "Milk", "Milk", "Milk", "Butter", "Butter", "Butter"],
                         "Person": ["Anna", "Anna", "Brian", "John", "John", "Carl", "Sarah", "Anna"],
                         "Sales":[200, 120, 340, 124, 243, 350, 500, 240],
                         "Quantity":[3, 5, 3, 8, 2, 7, 5, 4],
                         "Margin": [100, 20, 280, 50, 100, 67, 300, 200]})
df
```

```python
by_group = df.groupby(by= ["Product"])
by_group
```

```python
for Product, frame in by_group:
    print(f"Entries for '{Product}'")
    print("--------------------------------------------")
    print(f"{frame} \n\n")
```

```python
df.groupby(by=["Product"]).mean(numeric_only=True).round(2) #If you get a TypeError saying there are objects, then it's a pandas version issue. Use numeric_only=Trur when calling mean.
```
