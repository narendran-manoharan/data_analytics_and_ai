```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(2310)
df = pd.DataFrame(data={"SalesRep": ["John", "Julis", "Naren", "Chio", "Alex", "Naren", "Chio", "Peter", "Johannes"],
                        "Products": ["Mega", "LP", "AIA", "Intro", "MLOps", "RC", "Untitled", "Translation", "ML"],
                        "Sales": np.random.randint(4567, 9875, 9),
                        "Quantity": np.random.randint(479, 1000, 9),
                        "Margin": np.random.randint(25, 38, 9)})
df
```

```python
def salesforecast(x):
    return x.sum() *1.5
(df
.groupby(by=["Products"], observed=True)
.agg(Total_Sales = ("Sales", "sum"),
     Average_Sales = ("Sales", "mean"),
     Total_Quantity= ("Quantity", "sum"),
     Average_Quantity = ("Quantity", "mean"),
     Customer_Satisfaction = ("Margin", "prod"),
     Sellers = ("SalesRep", "nunique"),
     Team = ("SalesRep", "unique"),
     sales_forecast = ("Sales", salesforecast))
.reset_index()   
)
```

You wish to get the total sales per day and a count of the number of tables served per day and display this data in separate columns.
You should group by day
Create a column named Total_of_Sales, using the total_bill variable and the aggregation function 'sum'
Create a column named Tables_Served, using the time variable and the aggregation function 'count'
And you should reset the index

```python
(df
 .groupby(by=["day"], observed=True)
 .agg(total_sales = ("total_bill", "sum"),
      total_tables = ("time", "count")
      )
 .reset_index()

)
```
