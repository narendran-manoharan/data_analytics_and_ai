```python
def my_price_perception(price, cut):
    if cut in ["premium", "Ideal"] and price < 2000:
        return "worth it, buy now"
    elif price < 1000:
        return "buy"
    elif price < 2500:
        return "think twice"
    else:
        return "no, thanks"

df["my_price_perception"] = df.apply(lambda x: my_price_perception(x["price"], x["cut"]), axis = 1)
df
```

Each row of Weekend column should show True if the day value is Sat or Sun; else, it should show False. Use the apply function and write your lambda function within it.

```python
def weekday(day):
    if day in ["Sat", "Sun"]:
        return True
    else:
        return False

dfnew ["weekday"] = dfnew.apply(lambda x: weekday(x["day"]), axis = 1)
dfnew
```

### What this line does:
dfnew["weekend"] = dfnew.apply(lambda x: weekday(x["day"]), axis=1)
It adds a new column "weekend" to your DataFrame, where:

.apply(..., axis=1) means:
“For each row, run a function.”

lambda x: defines a small inline function that:

Takes a row x

Extracts the value from the "day" column of that row → x["day"]

Passes that value into your weekday() function

weekday(x["day"]) returns True if the value is "Sat" or "Sun", else False
