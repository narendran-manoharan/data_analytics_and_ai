```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.jointplot(data=df, x="total_bill", y="tip", hue="smoker")
plt.show()
```

```python
sns.jointplot(data=df, x="total_bill", y="tip", kind="hex")
plt.show()
```

### Lmplot - linear model plots
 - when there is a linear relationship between two variables

```python
df=pd.read_csv("insurance.csv")
df=df.sample(n=30, random_state=1)
df
```

```python
sns.lmplot(data=df, x="bmi", y="charges")
plt.show()
```

```python
sns.lmplot(data=df, x="age", y="charges")
plt.show()
```

```python
sns.lmplot(data= df, x="bmi", y="charges", col="smoker", ci=None)
plt.show()
```

```python
sns.lmplot(data= df, x="bmi", y="charges", col="smoker", row="region", ci=None)
plt.show()
```

```python
sns.lmplot(data=df, x="bmi", y="charges", hue="smoker", ci=None)
plt.show()
```

### Scatter takes all parameters like hue, size, and so on. 
- one unique thing about sns.scatter is that it allows numerical values to be added as hues.

```python
df=pd.read_csv("insurance.csv")
df=df.sample(n=30, random_state=1)
df
```

```python
fig, axes= plt.subplots(nrows=2, ncols=2, figsize=(12,8))
sns.scatterplot(data=df, x="age", y="charges", ax=axes[0,1])
plt.show()
```

```python
fig, axes= plt.subplots(nrows=2, ncols=2, figsize=(12,8))
sns.scatterplot(data=df, x="age", y="charges", ax=axes[0,1])
sns.scatterplot(data=df, x="age", y="region", ax=axes[1,0])
plt.show()
```

```python
fig, axes= plt.subplots(nrows=2, ncols=2, figsize=(12,8))
sns.scatterplot(data=df, x="age", y="charges", hue="bmi", ax=axes[0,0])
plt.show()
```

```python
fig, axes= plt.subplots(nrows=2, ncols=2, figsize=(12,8))
sns.scatterplot(data=df, x="age", y="charges", ax=axes[0,1], size="bmi")
plt.show()
```

```python
fig, axes= plt.subplots(nrows=2, ncols=2, figsize=(12,8))
sns.scatterplot(data=df, x="age", y="charges", ax=axes[0,1], size="bmi", sizes=(40, 200))
plt.show()
```
