```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("insurance.csv")
df
```

```python
df_corr = df.corr(method="pearson", numeric_only=True)
df_corr
```

```python
sns.heatmap(data=df_corr, annot=True, linewidth= 0.5, annot_kws={"size": 8}, cmap="viridis")
plt.show()
```

- dtype=np.bool ‚Üí makes all those zeros into Boolean False values.
- So now mask is a matrix of the same size as your correlation matrix, and all values are initially False.
- üí° Reminder: np.bool is deprecated ‚Äî you should use dtype=bool instead.

**mask[np.triu_indices_from(mask)] = True**
‚úÖ What it does:
- np.triu_indices_from(mask) gets the row and column indices of the upper triangle of the matrix (including the diagonal).
- Then it sets those positions in the mask array to True.
- The upper triangle is the top-right half of the matrix, which is symmetric to the bottom-left.
- This tells the heatmap:
üëâ ‚ÄúHide everything in the upper triangle.‚Äù

üß† Why we do this:
Correlation matrices are symmetric:

The correlation between A and B is the same as between B and A.

So we only need to show one triangle ‚Äî usually the lower one.
