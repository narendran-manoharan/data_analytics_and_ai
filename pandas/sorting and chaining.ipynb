```python
df_processed = (df
                .query("cut in ['Ideal', 'Premium']")
                .rename(mapper={"price":"investment potential"}, axis=1)
                .sort_values(by= ["carat", "cut", "color", "investment potential"], ascending=True)
                .filter(["carat", "cut", "investment potential"])
                .replace(to_replace={"Premium":"A1"})
                )
df_processed
```
