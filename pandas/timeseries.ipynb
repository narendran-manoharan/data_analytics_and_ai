```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(1223)
df = pd.DataFrame(data= {"Col1": np.random.randint(12, 80, 24)},
                    index= pd.date_range(start="17/07/2024", periods=24, freq= "h"))
df
```

```python
df ["hour"]= df.index.hour
df ["minute"] = df.index.minute
df ["seconds"] = df.index.second
df ["microseconds"] = df.index.microsecond
df ["day"] = df.index.day_name()
df ["weekday_number"] = df.index.weekday
df ["month_number"] = df.index.month
df ["month_name"] = df.index.month_name()
df ["year"] = df.index.year
df ["year_month"] = df.index.to_period("M")

df
```

```python
df = df.reset_index()
df
```

```python
np.random.seed(1457)
df2 = pd.DataFrame(data= {"Date of joining": pd.date_range(start= "08/06/2008", periods= 15, freq="ME"),
                          "Date of graduation": pd.date_range(start= "10/05/2011", periods=15, freq="D"),
                          "Grades": np.random.choice(["A", "B", "C"], size= 15)}) 
df2
```

```python
df2 ["Day"] = df2["Date of joining"].dt.day
df2
```

```python
df2["Month"] = df2["Date of graduation"].dt.month
df2
```

```python
df2["Month_name"] = df2["Date of graduation"].dt.month_name()
df2
```

```python
df2["Month-Year"] = df2["Date of graduation"].dt.to_period("M")
df2
```

```python
np.random.seed(1517)
df3 = pd.DataFrame(data= {"Name":np.random.choice(["Naren", "Julius", "Christian", "Harsha"], size=24),
                          "Project start": pd.date_range(start="11/08/2022", periods=24, freq="h")})
df3
```
