```python
df = sns.load_dataset("diamonds")
df.head(100)
df.head(10)
```

```python
def color_cut_selection(df, sort_variables):
    df = df.copy()
    df = (df
          .query("color in ['E']")
          .query("cut == 'Ideal' | cut == 'Premium'")
          .sort_values(by=sort_variables, ascending = False))
    return df
df.pipe(color_cut_selection, sort_variables=["color", "cut"])
```
