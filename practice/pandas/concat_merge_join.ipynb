### Merge

```python
df12 = pd.DataFrame(data= {"lkey": ["A1", "B1", "C1", "D1"],
                           "values": [33, 43, 53, 63]})

df13 = pd.DataFrame(data= {"rkey": ["A1", "B1", "C1", "DE1"],
                           "values": [318, 418, 518, 618]})
```

```python
pd.merge(df12, df13, how="right", on= "lkey")
```

```python
pd.merge(df12, df13, how= "right", on="values")
```

- in merge (talking from the dictionary perspective), merging happens only when:
        - the "values" are the same 
        - you merge left
        - you merge on "value"
 If you merge right, and the values are different, then the values from the right dictionary stays, but whatever is not part of left until values, turns into a NaN as seen above. 

```python
pd.merge(df12, df13, how="inner", on= "values")
```

```python
pd.merge(df12, df13, how="outer", on="values" )
```

```python
pd.merge(df12, df13, how="inner", left_on="lkey", right_on="rkey" )
```

```python
pd.merge(df12, df13, how="outer", left_on="lkey", right_on="rkey" )
```
