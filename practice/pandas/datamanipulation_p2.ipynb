```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(904)
df=pd.DataFrame(data= {"Col1": np.random.randint(0, 10, 10),
                       "Col2": np.random.choice(["A", "A.1", "B", "BP", "C", "ClassUnknown"], size=10),
                       "Superviser": np.random.choice(["Bob", "Alice", "Ken"], size=10),
                       "Quality":np.random.choice([True, False], size=10),
                       "Col5": np.random.randn(10),
                       "Outcome": np.random.choice(["Pass", "Fail", "Onhold"], size=10)},
                       index= pd.date_range(start="01/05/2005", periods=10, freq="W"))
df
```

```python
map_dict = {"BP": "BNegative", "C": "NeedsReview"}
map_dict
```

```python
df["Col2"].replace(to_replace=map_dict, inplace=True)
```

```python
df ["Col2"] = df["Col2"].replace(to_replace=map_dict)
df ["Col2"]
```
