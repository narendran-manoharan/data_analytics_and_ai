```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("flights")
df
```

```python
df["year"] = df["year"].astype("str")
df["month"] = df["month"].astype("str")
df["date"] = pd.to_datetime(df["month"]+ "-" +df["year"], format= "%b-%Y")
df = df.set_index("date")
df
```

```python
df.plot(
    kind= "line",
    y= "passengers",
    xlabel= "Date",
    ylabel= "Passengers",
    figsize= (10, 6),
    title = "Passengers X Time"

)
plt.show()
```

```python
(
    df
    .groupby(by=["year"], as_index=False)
    .agg(Average_levels = ("passengers", "mean"))
    .plot(kind="line", x="year", y="Average_levels", figsize= (5, 5))
)
plt.show()
```
