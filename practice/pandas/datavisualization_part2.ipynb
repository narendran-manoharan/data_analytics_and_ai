```python
df= sns.load_dataset("penguins")
df= df.sample(frac=0.5, random_state=1)
df.head()
```

```python
df["species"].value_counts()
```

```python
df.value_counts(["species"]).plot(kind = "bar", figsize=(10, 6))
plt.xticks(rotation=0)
plt.show()
```

```python
df.value_counts("island").plot(kind="barh", y="island", figsize=(10, 6))
plt.xticks(rotation=0)
plt.show()
```

```python
for col in df.select_dtypes(include="object").columns:
    df.value_counts(col).plot(kind="barh", y= "col", figsize=(10,6)),
    print(f"* {col} barh plot")
    plt.show()
    print("\n\n")
```

```python
df.groupby(["species", "island"]).size().unstack().plot(kind= "bar", figsize= (10, 6), stacked=False)
plt.xticks(rotation=0)
plt.show()
```

## You are interested in creating a stacked bar chart of categorical data grouped by day and time as you wish to easily see what times of the day are busiest at lunchtime or dinner time.

```python
df.groupby(["day", "time"], observed=False).size().unstack().plot(kind="bar", stacked=True, figsize = (10, 6))
plt.show()
```

```python
df= sns.load_dataset("penguins")
df
```

```python
df.value_counts("sex", normalize=True)
```

```python
df.value_counts("sex", normalize=True).plot(kind="pie", y="island", title= "Penguin sex pie chart")
plt.show()
```

```python
for col in df.select_dtypes(include="object").columns:
    df.value_counts(col, normalize=True).plot(kind="pie", y="", title=f"* {col} pie chart")
    plt.show()
    print("\n\n")
```

```python
df1= sns.load_dataset("tips")
df1= df1.head(50)
df1
```

```python
for col in df1.select_dtypes(include="category").columns:
    df1[col] = df1[col].astype("object")
    df1[col].value_counts(col, normalize=True).plot(kind= "pie", ylabel="", title= f"* {col} pie chart")
    plt.show()
    print("\n\n")
```
