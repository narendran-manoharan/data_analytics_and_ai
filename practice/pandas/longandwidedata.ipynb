```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(1)
df = pd.DataFrame(data= {"City": ["Dublin", "Dublin", "Dublin", "Dublin", "Corq", "Corq", "Corq", "Corq", "Galway", "Galway", "Galway", "Galway"],
                         "Day": [1, 2, 3, 4] *3,
                         "Temperature": np.random.choice([10, 25], size=12)})

df
```

```python
pd.pivot_table(data = df,
               index = ["City"],
               columns = ["Day"],
               values = "Temperature").reset_index().rename_axis(mapper=None, axis=1)
```

```python
df2 = pd.DataFrame(data= {"City": ["Dublin", "Corq", "Galway"],
                          "Day1": [9, 8, 2],
                          "Day2": [7, 6, 3],
                          "Day3": [10, 10, 6],
                          "Day4": [12, 10, 9]})
df2
```

```python
new_df2 = pd.melt(frame = df2,
        var_name = "Day",
        value_name = "Temperature",
        value_vars = ["Day1", "Day2", "Day3", "Day4"],
        id_vars = ["City"])
new_df2
```

```python
new_df2.sort_values("City", ascending=True)
```

```python
df3 = sns.load_dataset("diamonds")
df3
```

```python
filtered_df3 = df3.query("cut == 'Ideal' & color == 'E'")
filtered_df3
```

```python
new_table = pd.pivot_table(data = df3,
                           index = "cut",
                           columns= "color",
                           values= "price",
                           aggfunc= "mean")
new_table
```

```python
bnew_df3 = pd.pivot_table(data= df3,
               index= ["cut", "color"],
               columns= "clarity",
               values= "price",
               fill_value=0,
               aggfunc= "max")
bnew_df3
```

```python
bnew_df3.isnull().sum()
```

```python
bnew_df3 = bnew_df3.fillna(bnew_df3["IF"]).mean()
bnew_df3
```
