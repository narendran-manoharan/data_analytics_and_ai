```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel("Copy of UK MacroData.xlsx")
df.head()
```

```python
df.to_csv("UK MacroData.CSV")
df
```

```python
df["year"] = df["Date"].str[:4]
df["quarter"] = df["Date"].str[-2:]
df ["Date"] = pd.to_datetime(df["year"] + "-" + df["quarter"])
df= df.set_index("Date")
df.head()
```

```python
plt.plot(df["Gross Fixed Capital Formation (Investments)"])
plt.show()
```

```python
for col in df.select_dtypes(include="object").columns:
    print(f"{col} chart")
    print("\n\n")
    df[col].value_counts().plot(kind="line")
    plt.show()
```

```python
plt.plot(df[["Gross Fixed Capital Formation (Investments)", "GDP (£ m)"]])
plt.show()
```

```python
df = pd.read_excel("Online_retail.xlsx")
df
```

```python
fig, axes = plt.subplots(nrows=1 , ncols=2, figsize=(10,6))
axes[0].hist(data=df, x= "UnitPrice", bins=20)
axes[1].hist(data= df, x= "Quantity", bins=20)
plt.legend()
plt.show()
```

```python
for col in ["UnitPrice", "Quantity"]:
    plt.hist(data= df, x= col, label=col)
plt.legend()
plt.show()
```

```python
df = df.sample(n=50, random_state=9)
df.head(3)
```

```python
plt.boxplot(data= df, x="UnitPrice")
plt.show()
```

```python
list_of_data = []
col_boxplot = ["Quantity", "UnitPrice"]
for col in col_boxplot:
    list_of_data.append(df[col])
plt.boxplot(x= list_of_data)
plt.xticks(np.arange(1, len(col_boxplot)+1), col_boxplot)
plt.show()
```

```python
x, y= "UnitPrice", "Quantity"
plt.scatter(data= df, x= x, y=y, c=y, s=df[x]*25)
plt.ylabel(y)
plt.xlabel(x)
plt.show()
```
