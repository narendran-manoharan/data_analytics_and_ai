```python
df = pd.read_csv("Online_Retail.csv")
df = df.sample(n=50, random_state=9)
df.head(3)
```

### In Histogram, you pass through data, X, and bins.

```python
plt.hist(data= df, x="Quantity", bins=30)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,6))
axes[0,1].hist(data=df, x= "Quantity")
axes[1,1].hist(data=df["UnitPrice"], x= "Quantity")
plt.show()
```

```python
for col in ["Quantity","UnitPrice"]:
    plt.hist(data=df, x=col, label=col)
plt.legend()
plt.show()
```

### In Boxplot, you pass through data, and X.

```python
plt.boxplot(data= df, x="Quantity")
plt.show()
```

```python
list_of_data= []
varaibles_boxplot= ["Quantity", "UnitPrice"]
for col in varaibles_boxplot:
    list_of_data.append(df[col])

plt.boxplot(x=list_of_data)
plt.xticks(np.arange(1,len(varaibles_boxplot)+1), varaibles_boxplot)
plt.show()
```

```python
plt.boxplot(data=df, x="UnitPrice")
plt.show()
```

```python
plt.boxplot(data=df, x=df[["UnitPrice", "Quantity"]])
plt.show()
```

In Scatter plot, there's data, X, Y, c for color, and s for size

```python
x, y= "Quantity", "UnitPrice"
plt.scatter(data= df, x=x, y=y,c=y, s= df["UnitPrice"] * 3)
plt.ylabel(y)
plt.xlabel(x)
plt.show()
```

```python
df= pd.read_csv("insurance.csv")
df  
```

```python
categorical_count= df.value_counts("region")
categorical_count
```

```python
plt.bar(x=categorical_count.index, height= categorical_count)
plt.show()
```

```python
plt.barh(y= categorical_count.index, width= categorical_count)
plt.show()
```

```python
categorical_count= df.value_counts("region", normalize=True)
plt.pie(x= categorical_count, labels=categorical_count.index)
plt.title("Regional data")
plt.show()
```
