## Revision of missing data

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
df = pd.DataFrame(data= {"Product ID": np.random.choice([65457889, 987563487, 
                                                        np.nan], size= 4),
                         "Quantity": np.random.choice([189, 356, np.nan], 
                                                      size= 4),
                          "Customer reviews": np.random.choice([1, 5], size= 4),
                          "Sales": np.random.choice([1400, 2289, 4456, 5397], 
                                                    size= 4)                                                            
})
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df ["Quantity"] = df["Quantity"].fillna(df["Quantity"].mean())
df
```

```python
df.copy()
```

```python
np.random.seed(1)
df_2 = pd.DataFrame(data = {"Product ID": np.random.choice([345782, 745890,], size = 6),
                            "Quantity": np.random.choice([475, 1087, 1933, np.nan, np.nan], size = 6),
                            "Custom reviews": np.random.choice([500, 1000, np.nan], size = 6),
                            "Popularity": np.random.choice([1, 2, 3, 4, 5], size=6),
                            "Sales": np.random.choice([3645, 78541, np.nan], size = 6),
                            "Service class": np.random.choice([1, 2, np.nan], size = 6)})
df_2
```

## How to call the fillna function?
- Call out to col you want to work on. 
- Repeat the col name and open [] and call out once again the name of the column you're working on.
- then call out the .fillna() function
- inside the function, once again call you the col name [] you're working on
- finally, call out the statistic method you want to perform .mean(), .median(), .std() and close the bracket.

df["col name"] = df["col name"].fillna(df["col name"].mean())

```python
df_2["Custom reviews"] = df_2["Custom reviews"].fillna(df_2["Custom reviews"].std())
df_2
```

```python
df_2["Quantity"] = df_2["Quantity"].bfill().ffill()
df_2["Quantity"]
```
