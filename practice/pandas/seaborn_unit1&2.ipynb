```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df= pd.read_csv("insurance.csv").sample(n=50, random_state=9)
df["childre"] = df["children"].astype("str")
df.head(3)
```

```python
fig, axes= plt.subplots()
sns.scatterplot(df, x="age", y="charges", hue="smoker")
plt.title("Sns-Plt")
plt.xlabel("X-axis:\n Age")
plt.ylabel("Y-axis:\n Charges", rotation=None)
plt.legend(title="Legend")
plt.savefig("mytry2.png")
```

```python
fig, axes= plt.subplots(nrows= 3, ncols=3, figsize=(12,8))

sns.scatterplot(data=df, x="age", y="charges", hue="smoker", ax=axes[0, 1])
axes[0, 1].set_title("first plot")
sns.histplot(data=df, x="children", ax=axes[1,1])
axes[1, 1].set_title("second")
sns.barplot(data= df, x="sex", y="charges", ax=axes[2,0])

plt.tight_layout()
fig.suptitle("SNS", fontsize=12, y=1.1)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
