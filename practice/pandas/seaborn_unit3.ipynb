```python
df=sns.load_dataset("tips")
df
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
sns.kdeplot(data=df.filter(["total_bill", "tip"], axis=1), fill=True)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(12,8))

sns.kdeplot(data=df, ax=axes[0, 0], x="tip", hue="sex", fill=True, palette= "Set1")
sns.kdeplot(data=df, ax=axes[1, 1], x="tip", hue="time", fill=True, palette="Set2")
sns.kdeplot(data=df, ax=axes[2,0], x="tip", hue="day", fill=True, palette="Set3")
sns.kdeplot(data=df, ax=axes[3, 1], x="tip", hue="size", fill=True, palette="Set1")
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
sns.boxplot(data=df, x="size", y="tip", hue="time")
plt.show()
```

```python
sns.swarmplot(data=df, x="size" , y="tip", hue="time", dodge=True)
plt.show()
```

```python
fig, axes= plt.subplots(nrows=2, ncols=2, figsize=(12,8))
sns.boxplot(data=df, x="size", y="tip", hue="time", ax=axes[0,0])
sns.swarmplot(data=df, x="size", y="tip", hue="time", ax=axes[0, 1], dodge=True)
plt.show()
```
