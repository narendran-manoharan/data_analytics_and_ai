## Part 2 contains moving average and onwards

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(1733)
df = pd.DataFrame(data= {"Date": pd.date_range(start="10/10/2020", freq="d", periods=15),
                         "Product sold": np.random.randint(300, 1000, 15)})
df
```

```python
df["MovingAvg"] = df["Product sold"].rolling(window=4).mean()
df
```

```python
df["Move1"] = df["Product sold"].shift(periods=1)
df["Move2"] = df["Product sold"].shift(periods=-1)
df
```

```python
df
```

```python
df["Product sold diff"] = df["Product sold"].diff(periods=-1)
df
```

```python
df["Percentagechange"] = df["Product sold"].pct_change(periods=-1)
df
```
