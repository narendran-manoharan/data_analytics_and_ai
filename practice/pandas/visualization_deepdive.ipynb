```python
data = pd.DataFrame({"category": ["A", "B", "C"],
"value": [10, 25, 36]})

plt.bar(data["category"], data["value"], color="skyblue")
plt.title("Matplotlib bar chart")
plt.xlabel("category")
plt.ylabel("value")
plt.grid(axis = "y", linestyle= "--", alpha= 0.7)
plt.show()

sns.barplot(x="category", y="value", hue= "category", data=data, palette= "viridis", legend=False)
plt.title("SNS bar chart")
plt.grid(axis="y", linestyle = "--", alpha = 0.8)
plt.show()

fig= px.bar(data, x="category", y="value", title ="plotly example")
fig.show()
```
