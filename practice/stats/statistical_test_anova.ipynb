```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import pingouin as pg
sns.set_style("whitegrid")
```

Consider a dataset from pingouin datasets. It shows Pain threshold levels across different people, Hair color (Dark Brunette, Light Blond, Dark Blond, Light Brunette). The subject is the person's ID

```python
df1 = pg.read_dataset("anova")
df1
```

null: pain threshold is not normally distributed by hair color
alternative: pain threshold is normally distributed by hair color

```python
pg.normality(data=df1, dv="Pain threshold", group="Hair color", alpha=0.05)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14,6))
sns.boxplot(data=df1, x="Hair color", y="Pain threshold", ax=axes[0])
sns.swarmplot(data=df1, x="Hair color", y="Pain threshold", dodge=True, ax=axes[1])
plt.show()

```

```python
pg.anova(data=df1, dv="Pain threshold", between="Hair color", detailed=True)
```
